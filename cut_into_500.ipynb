```python
# 将pdf文件每五百页分割，返回文件名包含页码信息
import fitz


def split_pdf(input_pdf, output_prefix, chunk_size):
    input_doc = fitz.open(input_pdf)
    total_pages = input_doc.page_count

    for i in range(0, total_pages, chunk_size):
        output_doc = fitz.open()
        sub_range = range(i, min(i + chunk_size, total_pages))
        for page_num in sub_range:
            output_doc.insert_pdf(input_doc, from_page=page_num, to_page=page_num)

        output_file = (
            f"{output_prefix}_split_{sub_range.start + 1}-{sub_range.stop}.pdf"
        )
        output_doc.save(output_file)
        output_doc.close()


# 使用示例
input_pdf = "./data/pdf/二十四史全译.史记.第1册_text.pdf"
output_prefix = input_pdf.rsplit(".", maxsplit=1)[0]
chunk_size = 500
```
