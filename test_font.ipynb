```python
from fontTools.ttLib import TTFont

def get_supported_characters(font_path):
    try:
        font = TTFont(font_path)
        cmap_table = font['cmap']

        supported_codepoint = set()
        # supported_characters = set()
        # non_unicode_table = []

        for table in cmap_table.tables:
            if table.isUnicode():
                for codepoint, glyph_id in table.cmap.items():
                    supported_codepoint.add(codepoint)
                    # supported_characters.add(chr(codepoint))

            # else:
            #     non_unicode_table.append(table)

        return supported_codepoint

    except Exception as e:
        print(f"Error: {e}")
        return set(), []
```

```python
font_paths = ["./data/fonts/FangZhengFangSong-GBK-1.ttf","./data/fonts/FangZhengHeiTi-GBK-1.ttf","./data/fonts/FangZhengKaiTi-GBK-1.ttf","./data/fonts/FangZhengShuSong-GBK-1.ttf","./data/fonts/FangZhengZhunYuanFanTi.ttf"] # 替换为你的字体文件路径
supported_codepoints = [get_supported_characters(font_path) for font_path in font_paths]
```
